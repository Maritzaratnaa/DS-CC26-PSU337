Preprocessing

```python
def preprocess_text(text):

    if not isinstance(text, str) or not text.strip():
        return ""

    text = text.lower() # lowercase
    text = re.sub(r"http\S+|www\.\S+", " ", text) # hapus url
    text = re.sub(r"\S+@\S+", " ", text) # hapus email
    text = re.sub(r"[\\|(){}\[\]<>\"']", " ", text) # hapus simbol
    text = re.sub(r"\b\d+\b", " ", text) # hapus angka
    text = re.sub(r"[;:,\n\r\t]+", " ", text) # normalisasi separator
    text = re.sub(r"[!?*&%$@^=~`]", " ", text) # hapus simbol tambahan
    text = re.sub(r"\s+", " ", text).strip() # normalisasi spasi

    return text
```

```python
def remove_stopwords(text):

    tokens = text.split()

    filtered = []

    for token in tokens:
        if token in PROTECTED_TOKENS:
            filtered.append(token)
        elif token not in ALL_STOPWORDS and (
            len(token) > 2 or token in PROTECTED_TOKENS
        ):
            filtered.append(token)

    return " ".join(filtered)
```

```python
def full_preprocess(text):
    cleaned = preprocess_text(text)

    return remove_stopwords(cleaned)
```

```python
df["cleaned_text"] = df["deskripsi"].astype(str).apply(full_preprocess)

df[["deskripsi", "cleaned_text"]].head()
```

Extractor

```python
DICT_KEYS_SORTED = sorted(
    SKILL_DICT.keys(),
    key=lambda x: len(x),
    reverse=True
)

FUZZY_THRESHOLD = 85
FUZZY_MIN_LEN = 5
```

Helper

```python
def _normalize(text: str) -> str:
    return re.sub(r"\s+", " ", text.lower().strip())


def _dedupe_ordered(skills: List[str]) -> List[str]:
    seen = set()
    result = []
    for s in skills:
        key = s.lower()
        if key not in seen:
            seen.add(key)
            result.append(s)
    return result
```

Skill Extraction

```python
def extract_skills(raw_text: str, use_fuzzy: bool = True) -> List[str]:
    if not isinstance(raw_text, str) or not raw_text.strip():
        return []

    #lowercase, hapus noise
    text = preprocess_text(raw_text)
    found_skills = []
    consumed_spans = []

    #Regex matching
    temp_text = text
    for pattern, canonical in REGEX_PATTERNS:
        for m in re.finditer(pattern, temp_text, re.IGNORECASE):
            found_skills.append(canonical)
            placeholder = " " * len(m.group())
            temp_text = temp_text[:m.start()] + placeholder + temp_text[m.end():]

    working_text = temp_text

    #Dictionary matching
    for key in DICT_KEYS_SORTED:
        if len(key) < 2:
            continue

        escaped = re.escape(key)
        pattern = r"(?<!\w)" + escaped + r"(?!\w)"
        for m in re.finditer(pattern, working_text, re.IGNORECASE):
            found_skills.append(SKILL_DICT[key])
            placeholder = " " * len(m.group())
            working_text = working_text[:m.start()] + placeholder + working_text[m.end():]

    #Fuzzy matching
    if use_fuzzy:
        tokens = [t for t in working_text.split() if len(t) >= FUZZY_MIN_LEN]
        unique_tokens = list(set(tokens))

        for token in unique_tokens:
            result = process.extractOne(
                token,
                DICT_KEYS_SORTED,
                scorer=fuzz.ratio,
                score_cutoff=FUZZY_THRESHOLD,
            )
            if result:
                matched_key, score, _ = result
                if not re.fullmatch(r"\d+", token):
                    found_skills.append(SKILL_DICT[matched_key])

    #Deduplikasi
    return _dedupe_ordered(found_skills)
```
